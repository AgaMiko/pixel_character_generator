```python
import os
from datetime import datetime
import time
import random
random.seed(datetime.now())
```

```python
from PIL import Image

DIR ='C:\\Users\\1234\\OneDrive - pg.edu.pl\\Bazy danych\\rpg_char'

spreatsheets=[]
# r=root, d=directories, f = files
for r, d, f in os.walk(DIR):
    for file in f:
        if file.endswith(".png"):
            spreatsheets.append(os.path.join(r, file))
print(spreatsheets)       
```

```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

start_h = 260
start_w = 0
char_h = 64
char_w = 64
no_chars = 4
for i,spreatsheet in enumerate(spreatsheets):
    for nr in range(no_chars):
        image=mpimg.imread(spreatsheet)
        resized_im = image[start_h+char_h*nr:start_h+char_h*(nr+1),start_w:start_w+char_w,:]
        #imgplot = plt.imshow(resized_im)
        
        path = 'data\\'+str(nr)+'\\'
        if not os.path.exists(path):
            os.makedirs(path)
            
        plt.imsave(path+str(i)+'.png', resized_im)
```
